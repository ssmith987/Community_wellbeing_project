```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
median_income_forecast = pd.read_csv("CSV files/median_income_estimate_by_zip.csv")
median_income_df = pd.DataFrame(median_income_forecast)
```

```python
median_income_df.head()
```

```python
median_home_price = pd.read_csv("ZipCode_File/Zip_Zhvi_AllHomes.csv")
```

```python
median_home_price_df = pd.DataFrame(median_home_price)

median_home_price_df.head()
```

```python
med_home_price_df = median_home_price_df.rename(columns={'RegionName':'zip'})

med_home_price_df
```

```python
median_price_income = pd.merge(median_income_df, med_home_price_df, how='inner', on='zip')
```

```python
median_price_income
```

```python
median_price_income.plot(kind='scatter', x='median_inc', y="2016-12", grid=True)
plt.ylabel('Median Home Price ($)')
plt.xlabel('Median Income ($)')
plt.title('Median Home Price vs Median Income ($)')
plt.plot()
```

```python
plt.scatter(x=median_price_income['median_inc'], y=median_price_income['2016-12'], c=median_price_income['zip'])
```

```python
median_price_income.plot(kind='bar', x='City', y="median_inc", grid=True)
```

```python
city_grp = median_price_income.groupby(['City'], as_index=False).median()
```

```python
print(city_grp['median_inc'].median())
print(city_grp['2016-12'].median())
```

```python
top_city_inc = city_grp.nlargest(3,'median_inc')
top_city_inc

bottom_city_inc = city_grp.nsmallest(3,'median_inc')
bottom_city_inc

t_b_city_inc = top_city_inc.append(bottom_city_inc)

t_b_city_inc
```

```python
t_b_city_inc.plot(kind='bar', x='City', y="median_inc")
t_b_city_inc.plot(kind='line', x='City', y='2016-12')

plt.plot()
```

```python
city_grp['inc_price_comp'] = city_grp['median_inc']/city_grp['2016-12']*100
city_grp
```

```python
top_comp = city_grp.nlargest(3,'inc_price_comp')
top_comp

bottom_comp = city_grp.nsmallest(3,'inc_price_comp')
bottom_comp

t_b_comp = top_comp.append(bottom_comp)

t_b_comp
```
