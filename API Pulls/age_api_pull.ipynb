```python
import pandas as pd
import requests
import json
```

```python
url_zip = "http://datasurfer.sandag.org/api/estimate/2016/zip"

list_zip = requests.get(url_zip).json()

print(json.dumps(list_zip, indent=4, sort_keys=True))
```

```python
zip_codes = pd.DataFrame(list_zip)
zip_codes.head()
```

```python
age_url = "http://datasurfer.sandag.org/api/forecast/13/zip/92101/age"

age_request = requests.get(age_url).json()
print(json.dumps(age_request, indent=4, sort_keys=True))
```

```python
age_df = pd.DataFrame(age_request)
age_df
```

```python
age_df = age_df.iloc[0:0]
age_df
```

```python
for code in zip_codes['zip']:
    age_url = "http://datasurfer.sandag.org/api/forecast/13/zip/" + code + "/age"
    age_request = requests.get(age_url).json()
    
    age_df = age_df.append(age_request, ignore_index=True)
```
