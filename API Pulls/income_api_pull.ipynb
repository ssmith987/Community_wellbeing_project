```python
import pandas as pd
import requests
import json
```

```python
url_zip = "http://datasurfer.sandag.org/api/estimate/2016/zip"

list_zip = requests.get(url_zip).json()

print(json.dumps(list_zip, indent=4, sort_keys=True))
```

```python
zip_codes = pd.DataFrame(list_zip)
zip_codes.head()
```

```python
income_url = "http://datasurfer.sandag.org/api/forecast/13/zip/91901/income"
income_request = requests.get(income_url).json()

print(json.dumps(income_request, indent=4, sort_keys=True))
```

```python
income_df = pd.DataFrame(income_request)
income_df
```

```python
income_df = income_df.iloc[0:0]
income_df
```

```python
for code in zip_codes['zip']:
    income_url = "http://datasurfer.sandag.org/api/forecast/13/zip/" + code + "/income"
    income_request = requests.get(income_url).json()
    
    income_df = income_df.append(income_request, ignore_index=True)
```

```python
income_df.count()
```

```python
income_df.to_csv("income_estimate_by_zip.csv")
```
