```python
import pandas as pd
import requests
import json
```

```python
url_zip = "http://datasurfer.sandag.org/api/estimate/2016/zip"

list_zip = requests.get(url_zip).json()

print(json.dumps(list_zip, indent=4, sort_keys=True))
```

```python
zip_codes = pd.DataFrame(list_zip)
zip_codes.head()
```

```python
median_income_url = "http://datasurfer.sandag.org/api/estimate/2016/zip/92020/income/median"
median_income_request = requests.get(median_income_url).json()

print(json.dumps(median_income_request, indent=4, sort_keys=True))
```

```python
median_income_df = pd.DataFrame(median_income_request)
median_income_df
```

```python
median_income_df = median_income_df.iloc[0:0]
median_income_df
```

```python
for code in zip_codes['zip']:
    try:
        median_income_url = "http://datasurfer.sandag.org/api/estimate/2016/zip/" + code + "/income/median"
        median_income_request = requests.get(median_income_url).json()
    
        median_income_df = median_income_df.append(median_income_request, ignore_index=True)
        
    except:
        continue
```

```python
median_income_df.count()
```

```python
median_income_df.to_csv("median_income_estimate_by_zip.csv")
```
