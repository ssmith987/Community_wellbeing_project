```python
import pandas as pd
import requests
import json
```

```python
url_zip = "http://datasurfer.sandag.org/api/estimate/2016/zip"

list_zip = requests.get(url_zip).json()

print(json.dumps(list_zip, indent=4, sort_keys=True))
```

```python
zip_codes = pd.DataFrame(list_zip)
zip_codes.head()
```

```python
jobs_url = "http://datasurfer.sandag.org/api/forecast/13/zip/91901/jobs"
jobs_request = requests.get(jobs_url).json()

print(json.dumps(jobs_request, indent=4, sort_keys=True))
```

```python
jobs_df = pd.DataFrame(jobs_request)
jobs_df
```

```python
jobs_df = jobs_df.iloc[0:0]
jobs_df
```

```python
for code in zip_codes['zip']:
    jobs_url = "http://datasurfer.sandag.org/api/forecast/13/zip/" + code + "/jobs"
    jobs_request = requests.get(jobs_url).json()
    
    jobs_df = jobs_df.append(jobs_request, ignore_index=True)
```

```python
jobs_df.count()
```

```python
jobs_df.to_csv("job_estimate_by_zip.csv")
```
