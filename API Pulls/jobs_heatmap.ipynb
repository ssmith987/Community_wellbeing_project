```python
import numpy as np
import pandas as pd
from api_keys import api_key
import gmaps
```

```python
job_data = pd.read_csv("../CSV Files/job_estimate_by_zip.csv")
```

```python
census_locs = pd.read_csv("../2018_Gaz_zcta_national copy.csv", sep='\t')
census_locs = census_locs.rename(columns={'GEOID':'zip'})
census_locs.columns
```

```python
job_locs_df = pd.merge(job_data, census_locs, how='inner')
job_locs_df = job_locs_df.loc[job_locs_df['category']=='Total Jobs']
job_locs_df.head()
```

```python
job_df = job_locs_df.groupby('zip').sum()
job_data_group = job_df.reset_index()
job_data_group = job_data_group[['zip', 'jobs']]
job_data_group

locs_df = job_locs_df.groupby('zip').max()
locs_df = locs_df.reset_index()

job_data = pd.merge(job_data_group, locs_df, how='left', on='zip')
job_data['jobs_x'].sum()
```

```python
gmaps.configure(api_key=api_key)
```

```python
locations = job_data[['INTPTLAT','INTPTLONG']].astype(float)
job_numbers = job_data['jobs_x'].astype(float)
```

```python
fig = gmaps.figure()

heat_layer = gmaps.heatmap_layer(locations, weights=job_numbers, 
                                 dissipating=False,
                                 point_radius = .05)
```

```python
heat_layer.dissipating = False
#heat_layer.max_intensity = 500000
heat_layer.point_radius = .05

fig.add_layer(heat_layer)

fig
```

```python
vacancy_rate_data = pd.read_csv("../CSV Files/housing_estimate_by_zip.csv")
vacancy_rate_data = vacancy_rate_data.loc[vacancy_rate_data['year']==2020]
vacancy_rate_data = vacancy_rate_data.loc[vacancy_rate_data['unit_type']=='Total Units']
vacancy_rate_data
```

```python
vacancy_loc = pd.merge(vacancy_rate_data, census_locs, on='zip', how='left')
vacancy_loc = vacancy_loc.dropna()
vacancy_loc[vacancy_loc['vacancy_rate'] < 0] = 0
```

```python
vacancy_locations = vacancy_loc[['INTPTLAT','INTPTLONG']].astype(float)
vacancy_data = vacancy_loc['vacancy_rate']
```

```python
vacancy_fig = gmaps.figure()

vacancy_layer = gmaps.heatmap_layer(vacancy_locations, weights=vacancy_data, 
                                 dissipating=False,
                                 point_radius = .075)

vacancy_fig.add_layer(vacancy_layer)

vacancy_fig
```
