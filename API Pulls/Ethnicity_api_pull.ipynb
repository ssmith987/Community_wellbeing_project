```python
import pandas as pd
import requests
import json
```

```python


url_zip = "http://datasurfer.sandag.org/api/estimate/2016/zip"

list_zip = requests.get(url_zip).json()

print(json.dumps(list_zip, indent=4, sort_keys=True))
```

```python
zip_codes = pd.DataFrame(list_zip)
```

```python
zip_codes.head()
```

```python
ethnicity_url = "http://datasurfer.sandag.org/api/forecast/12/zip/92020/ethnicity/change"

#ethnicity_data = []

#for code in zip_codes['zip']:
   # ethnicity_data = requests.get(ethnicity_url.json()


list_zip = requests.get(ethnicity_url).json()

print(json.dumps(list_zip, indent=4, sort_keys=True))
```

```python
ethnicity_df_test = pd.DataFrame(list_zip)
ethnicity_df_test
```

```python
for code in zip_codes['zip']:
    ethnicity_url = "http://datasurfer.sandag.org/api/forecast/12/zip/" + code + "/ethnicity/change"
    ethnicity_request = requests.get(ethnicity_url).json()
    
    ethnicity_df_test = ethnicity_df_test.append(ethnicity_request, ignore_index=True)
    
```

```python
ethnicity_df_test
```

```python
ethnicity_df_test.to_csv("ethnicity_by_zip")
```
