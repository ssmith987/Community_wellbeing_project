```python
import pandas as pd
import requests
import json
```

```python
url_zip = "http://datasurfer.sandag.org/api/estimate/2016/zip"

list_zip = requests.get(url_zip).json()

print(json.dumps(list_zip, indent=4, sort_keys=True))
```

```python
zip_codes = pd.DataFrame(list_zip)
zip_codes.head()
```

```python
housing_url = "http://datasurfer.sandag.org/api/forecast/13/zip/91901/housing"
housing_request = requests.get(housing_url).json()

print(json.dumps(housing_request, indent=4, sort_keys=True))
```

```python
housing_df = pd.DataFrame(housing_request)
housing_df
```

```python
housing_df = housing_df.iloc[0:0]
housing_df
```

```python
for code in zip_codes['zip']:
    housing_url = "http://datasurfer.sandag.org/api/forecast/13/zip/" + code + "/housing"
    housing_request = requests.get(housing_url).json()
    
    housing_df = housing_df.append(housing_request, ignore_index=True)
```

```python
housing_df.count()
```

```python
housing_df.to_csv("housing_estimate_by_zip.csv")
```
